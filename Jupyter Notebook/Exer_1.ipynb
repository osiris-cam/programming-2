```python
import matplotlib.pyplot as plt

import pandas as pd
wktime1 = pd.read_csv('wktime.T001.his', delimiter= ' ', header= None) #Open the file and save it in the variable as dataframe
plt.plot(wktime1[0], wktime1[1], color= 'c', marker= 'o', label ='Wktime1') #label of my first line and X and Y is defined

wktime4 = pd.read_csv('wktime.T004.his', delimiter= ' ', header= None) #Open the file and save it in the variable as dataframe
plt.plot(wktime4[0], wktime4[1], color= 'b', marker= '*', label ='Wktime4') #label of my first line and X and Y is defined
plt.xlabel ('W') #X label
plt.ylabel('Probability p(w)') # Y label
plt.title("Probability distribution") # Plot title

plt.xscale("log") #Logarithmic scale
plt.yscale("log")
plt.xlim(0.001, 100) #Range in X
plt.ylim(0.000001, 100) #Range in  Y
plt.show()
```
